### The command for cutting video in 10 mins

```python

input_video_path = "/home/ipd/Scripts/SAM2/videos/9953_Act.mp4"

output_video_path = "/home/ipd/Scripts/SAM2/videos/9953_Act.mp4"

number_of_splits = 3

start_hours = 0
start_minutes = 0
start_seconds = 26

for i in range (number_of_splits):
    # target is 00:10:26
    start_time = f"{start_hours:02}:{start_minutes:02}:{start_seconds:02}"
    index = i + 1
    start_minutes += 10
    end_time = f"{start_hours:02}:{start_minutes:02}:{start_seconds:02}"
    ffmpeg_command = [
        "ffmpeg",
        "-i", input_video_path,
        "-ss", start_time,
        "-to", end_time,
        "-c:v", "copy",
        "-c:a", "copy",
       output_video_path.replace(".mp4", '') + "_0" + str(index) + ".mp4", 
    ]
    subprocess.run(ffmpeg_command)
```
